```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\shash\\Downloads\\datasets\\Amazon Sale Report.csv")
df
```

```python
df.dtypes
```

```python
# Drop duplicates
df.drop_duplicates(inplace=True)
# Check for missing values
print("\nMissing values:")
print(df.isnull().sum())
```

```python
# Convert Date to datetime
df["Date"] = pd.to_datetime(df["Date"], errors='coerce')
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder

# List of object columns to convert to integers
obj_cols = [
    "Status",
    "Fulfilment",
    "ship-service-level",
    "Category",
    "Size",
    "Courier Status",
    "currency",
    "ship-city",
    "ship-state"
]

# Apply Label Encoding
le = LabelEncoder()
for col in obj_cols:
    df[col] = le.fit_transform(df[col])
```

```python
df["B2B"] = df["B2B"].astype(int)
```

```python
df.dtypes
```

```python
scaler = StandardScaler()
df[["Qty", "Amount"]] = scaler.fit_transform(df[["Qty", "Amount"]])

# ✅ Step 5: Save Transformed Data
df.to_csv("processed_chocolate_sales.csv", index=False)
print("✅ Processed data saved as 'processed_chocolate_sales.csv'.")
```
